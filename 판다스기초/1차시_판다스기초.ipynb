## 💡 Pandas란 무엇인가요?

> **Pandas = 파이썬에서 '표(엑셀)'를 다루는 도구**

- 엑셀로 하는 거의 모든 작업을 코드로 할 수 있습니다
- 수십만 행의 데이터도 빠르게 처리할 수 있습니다
- 데이터 분석가, 데이터 과학자들이 가장 많이 사용하는 라이브러리입니다

```
엑셀 시트 한 장 = DataFrame (데이터프레임)
엑셀의 한 열   = Series (시리즈)
```

## 📦 Step 1. 라이브러리 불러오기

> 도구를 사용하기 전에 먼저 '불러와야' 합니다. 마치 앱을 실행하는 것처럼요!

````python
# pandas 라이브러리를 불러옵니다
# 'as pd'는 앞으로 pandas를 pd라는 짧은 이름으로 부르겠다는 뜻입니다
import pandas as pd

# numpy는 숫자 계산을 도와주는 라이브러리입니다
import numpy as np

print("✅ 라이브러리 불러오기 완료!")
print(f"Pandas 버전: {pd.__version__}")
````

---

## 📌 Step 2. Series - 판다스의 기본 단위

> **Series = 엑셀의 '한 열'**  
> 값들의 목록 + 각 값에 붙는 번호(인덱스)로 이루어져 있습니다.

```
인덱스  |  값
  0    |  85
  1    |  92
  2    |  78
  3    |  95
```

````python
# 학생들의 수학 점수 목록으로 Series를 만들어봅니다
# 리스트([ ])를 pd.Series()로 감싸면 됩니다
math_scores = pd.Series([85, 92, 78, 95, 88])

print("=== 수학 점수 Series ===")
print(math_scores)
print()
print(f"데이터 타입: {type(math_scores)}")
````

````python
# 인덱스를 숫자 대신 이름으로 지정할 수도 있습니다
# index= 옵션으로 각 값에 이름을 붙여줍니다
math_scores_named = pd.Series(
    [85, 92, 78, 95, 88],
    index=['김민준', '이서연', '박지호', '최예린', '정도윤']  # 학생 이름을 인덱스로
)

print("=== 이름이 있는 수학 점수 ===")
print(math_scores_named)
print()

# 특정 학생 점수 조회
print(f"이서연 점수: {math_scores_named['이서연']}점")

# 기본 통계도 바로 계산 가능!
print(f"평균 점수: {math_scores_named.mean():.1f}점")
print(f"최고 점수: {math_scores_named.max()}점")
````

---

## 📌 Step 3. DataFrame - 판다스의 핵심!

> **DataFrame = 엑셀 시트 한 장 = 행과 열로 이루어진 표**  
> 여러 개의 Series를 모아놓은 것이라고 생각하면 됩니다.

```
이름   | 국어 | 수학 | 영어 | 학년
김민준 |  85  |  92  |  88  |  2
이서연 |  92  |  78  |  95  |  1
  ...  |  ... |  ... |  ... | ...
```

````python
# 딕셔너리(사전) 형태로 데이터를 만들고 DataFrame으로 변환합니다
# 딕셔너리의 '키' = 열(컬럼) 이름
# 딕셔너리의 '값' = 각 열의 데이터 목록

data = {
    '이름':  ['김민준', '이서연', '박지호', '최예린', '정도윤', '한지수', '오승현', '윤채원'],
    '학년':  [2, 1, 3, 2, 1, 3, 2, 1],
    '국어':  [85, 92, 78, 95, 88, 76, 91, 83],
    '수학':  [92, 78, 95, 88, 72, 85, 79, 96],
    '영어':  [88, 95, 82, 91, 85, 90, 77, 89],
    '과학':  [79, 88, 91, 85, 93, 82, 88, 75]
}

# 딕셔너리를 DataFrame으로 변환!
df = pd.DataFrame(data)

print("=== 학생 성적 데이터 ===")
df  # Jupyter에서는 print() 없이도 표 형태로 예쁘게 출력됩니다
````

---

## 📌 Step 4. 데이터 기본 확인 함수들

> 데이터를 처음 받으면 '어떻게 생겼는지' 파악하는 것이 가장 먼저입니다!
> 마치 새로운 책을 받으면 목차부터 훑어보는 것처럼요.

````python
# head() - 앞에서 몇 줄만 보기 (기본: 5줄)
# 데이터가 너무 많을 때, 처음 몇 개만 확인할 때 사용합니다
print("=== 처음 3줄 확인 (head) ===")
df.head(3)
````

````python
# tail() - 뒤에서 몇 줄 보기
print("=== 마지막 3줄 확인 (tail) ===")
df.tail(3)
````

````python
# shape - 데이터의 크기 (행의 수, 열의 수)
print(f"데이터 크기 (행, 열): {df.shape}")
print(f"→ {df.shape[0]}명의 학생, {df.shape[1]}개의 항목")
````

````python
# info() - 데이터 전체 정보 요약
# 각 열의 이름, 데이터 개수, 타입을 한눈에 볼 수 있습니다
print("=== 데이터 정보 요약 (info) ===")
df.info()
````

````python
# describe() - 숫자 데이터의 통계 요약
# count(개수), mean(평균), std(표준편차), min(최소), max(최대) 등을 한번에!
print("=== 통계 요약 (describe) ===")
df.describe()
````

````python
# columns - 열 이름 목록 확인
print(f"컬럼(열) 목록: {list(df.columns)}")

# dtypes - 각 열의 데이터 타입 확인
print()
print("각 열의 데이터 타입:")
print(df.dtypes)
````

## 📌 Step 5. 데이터 선택하기

### 5-1. 열(컬럼) 선택
> 원하는 열만 골라서 볼 수 있습니다!

````python
# 열 하나 선택 - 대괄호 [] 안에 열 이름을 씁니다
# 결과는 Series 형태로 나옵니다
print("=== 이름 열만 선택 ===")
print(df['이름'])
print()

# 수학 점수만 선택
print("=== 수학 점수 열만 선택 ===")
print(df['수학'])
````

````python
# 여러 열 선택 - 대괄호 안에 열 이름 리스트를 씁니다 [[ ]]
# 결과는 DataFrame 형태로 나옵니다
print("=== 이름, 수학, 영어 열만 선택 ===")
df[['이름', '수학', '영어']]
````

### 5-2. 행(row) 선택 - loc와 iloc

> - **loc** = 인덱스 **이름**으로 선택 (label 기반)
> - **iloc** = 인덱스 **번호**로 선택 (position 기반, i = integer)

```
비유: 친구를 찾을 때
loc  → 이름으로 찾기 ("김민준 어디있어?")
iloc → 번호로 찾기  ("3번 자리 학생 어디있어?")
```

````python
# iloc - 번호(위치)로 행 선택
# 파이썬은 0부터 시작합니다!
print("=== 첫 번째 행 (iloc[0]) ===")
print(df.iloc[0])  # 첫 번째 학생
print()

print("=== 두 번째 ~ 네 번째 행 (iloc[1:4]) ===")
df.iloc[1:4]  # 1번, 2번, 3번 인덱스 (4번 미포함!)
````

````python
# iloc으로 행과 열 동시에 선택
# df.iloc[행 번호, 열 번호]
print("=== 1~3번째 행의 이름, 수학 열만 ===")
df.iloc[0:3, 0:3]  # 0~2번 행, 0~2번 열
````

````python
# loc - 인덱스 이름으로 행 선택
# 우리 df는 인덱스가 숫자이므로 숫자를 씁니다
print("=== loc으로 0번 행 선택 ===")
print(df.loc[0])
print()

# loc으로 특정 행과 특정 열 선택
print("=== 0~2번 행의 이름, 국어, 수학 열만 ===")
df.loc[0:2, ['이름', '국어', '수학']]  # loc은 끝 번호 포함!
````

## 📌 Step 6. CSV 파일 불러오기

> 실제 데이터 분석에서는 파일을 불러와서 사용합니다.

## 📚 1차시 핵심 정리

| 개념 | 설명 | 코드 |
|------|------|------|
| **Series** | 1차원 데이터 (한 열) | `pd.Series([1, 2, 3])` |
| **DataFrame** | 2차원 표 데이터 | `pd.DataFrame(data)` |
| **head()** | 앞 n줄 보기 | `df.head(5)` |

| **info()** | 데이터 정보 요약 | `df.info()` |
| **describe()** | 통계 요약 | `df.describe()` |
| **열 선택** | 특정 열 선택 | `df['열이름']` |
| **iloc** | 번호로 행 선택 | `df.iloc[0]` |
| **loc** | 이름으로 행 선택 | `df.loc[0, '이름']` |
| **read_csv** | CSV 파일 불러오기 | `pd.read_csv('파일.csv')` |
