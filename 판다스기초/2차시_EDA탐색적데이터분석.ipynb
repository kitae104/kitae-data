## 💡 EDA란?

> **EDA (Exploratory Data Analysis) = 탐색적 데이터 분석**  
> 분석을 시작하기 전에 데이터를 충분히 들여다보고, 이상한 점을 찾아 정리하는 과정

```
요리를 시작하기 전에 재료를 꺼내서 상태 확인하기 = EDA
🥕 신선한가? (결측값 확인)
🥕 중복된 재료는 없나? (중복 확인)
🥕 필요한 재료만 선별하기 (필터링)
```

````python
# 필요한 라이브러리를 불러옵니다
import pandas as pd
import numpy as np

print("✅ 라이브러리 불러오기 완료!")
````

---

## 📌 Step 1. 실습 데이터 준비

> 이번에는 현실적인 데이터처럼 **결측값과 중복이 포함된** 데이터를 만들어봅니다.

````python
# 결측값(NaN)과 중복이 포함된 학생 성적 데이터
# NaN = Not a Number, 즉 '값이 없음'을 나타냅니다
# np.nan 으로 결측값을 표현합니다

data = {
    '이름':  ['김민준', '이서연', '박지호', '최예린', '정도윤', 
             '한지수', '오승현', '윤채원', '이서연', '박지호'],  # 이서연, 박지호 중복!
    '학년':  [2, 1, 3, 2, 1, 3, 2, 1, 1, 3],
    '국어':  [85, 92, 78, 95, 88, np.nan, 91, 83, 92, 78],   # 한지수 결측!
    '수학':  [92, 78, np.nan, 88, 72, 85, 79, 96, 78, np.nan], # 박지호 결측!
    '영어':  [88, 95, 82, np.nan, 85, 90, 77, 89, 95, 82],   # 최예린 결측!
    '과학':  [79, 88, 91, 85, 93, 82, np.nan, 75, 88, 91]    # 오승현 결측!
}

df = pd.DataFrame(data)

print("=== 원본 데이터 (결측값, 중복 포함) ===")
df
````

---

## 📌 Step 2. 결측값(Missing Value) 처리

> **결측값** = 데이터가 없는 칸 (엑셀의 빈 셀과 같습니다)  
> 결측값이 있으면 계산이 틀리거나 오류가 날 수 있어서 반드시 처리해야 합니다!

````python
# isnull() - 각 값이 결측값인지 확인 (True = 결측값)
print("=== 결측값 여부 확인 (True = 결측값) ===")
df.isnull()
````

````python
# isnull().sum() - 각 열마다 결측값이 몇 개인지 합산
print("=== 열별 결측값 개수 ===")
missing_count = df.isnull().sum()
print(missing_count)
print()

# 결측값 비율도 계산해봅니다
print("=== 열별 결측값 비율 ===")
missing_ratio = (df.isnull().sum() / len(df) * 100).round(1)
for col, ratio in missing_ratio.items():
    print(f"  {col}: {ratio}%")
````

### 결측값 처리 방법 3가지

```
방법 1: 그냥 삭제 (dropna)  → 결측값이 있는 행/열 자체를 제거
방법 2: 평균으로 채우기 (fillna)  → 평균, 중앙값 등으로 대체
방법 3: 특정 값으로 채우기 (fillna)  → 0이나 '없음' 등으로 대체
```

````python
# 방법 1: dropna() - 결측값이 하나라도 있는 행을 삭제
df_dropped = df.dropna()

print(f"원본 행 수: {len(df)}")
print(f"결측 행 삭제 후: {len(df_dropped)}")
print()
print("=== 결측값 행 삭제 후 데이터 ===")
df_dropped
````

````python
# 방법 2: fillna(평균값) - 결측값을 각 과목 평균으로 채우기
# 원본 데이터를 복사해서 사용합니다 (원본을 보존하기 위해)
df_filled = df.copy()

# 국어, 수학, 영어, 과학 점수의 결측값을 각 과목 평균으로 채웁니다
for col in ['국어', '수학', '영어', '과학']:
    col_mean = df_filled[col].mean()  # 해당 열의 평균
    df_filled[col] = df_filled[col].fillna(col_mean)  # 결측값을 평균으로 대체
    df_filled[col] = df_filled[col].round(1)  # 소수점 1자리로 반올림
    print(f"{col} 평균: {col_mean:.1f}점으로 결측값 대체")

print()
print("=== 결측값을 평균으로 채운 데이터 ===")
df_filled
````

````python
# 결측값이 잘 처리됐는지 확인
print("=== 처리 후 결측값 확인 ===")
print(df_filled.isnull().sum())
````

---

## 📌 Step 3. 중복 데이터 처리

````python
# duplicated() - 중복된 행을 확인합니다 (True = 중복)
print("=== 중복 여부 확인 ===")
print(df.duplicated())
print()
print(f"중복 행 수: {df.duplicated().sum()}개")
````

````python
# 어떤 행이 중복인지 확인
print("=== 중복된 행 내용 ===")
df[df.duplicated()]
````

````python
# drop_duplicates() - 중복 행 제거
df_no_dup = df_filled.drop_duplicates()
df_no_dup = df_no_dup.reset_index(drop=True)  # 인덱스를 0부터 새로 정렬

print(f"중복 제거 전: {len(df_filled)}행")
print(f"중복 제거 후: {len(df_no_dup)}행")
print()
print("=== 중복 제거 완료 ===")
df_no_dup
````

---

## 📌 Step 4. 새로운 열 추가 및 데이터 변환

````python
# 정제된 데이터로 작업합니다
df_clean = df_no_dup.copy()

# 새로운 열 추가: 총점 계산
df_clean['총점'] = df_clean['국어'] + df_clean['수학'] + df_clean['영어'] + df_clean['과학']

# 새로운 열 추가: 평균 계산 (소수점 1자리)
df_clean['평균'] = df_clean['총점'] / 4
df_clean['평균'] = df_clean['평균'].round(1)

print("=== 총점, 평균 추가 완료 ===")
df_clean
````

````python
# 조건에 따라 등급 매기기
# 함수를 만들어서 apply()로 적용합니다

def get_grade(score):
    """평균 점수에 따라 등급을 반환하는 함수"""
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    else:
        return 'D'

# apply() = 각 행/열에 함수를 적용하는 메서드
df_clean['등급'] = df_clean['평균'].apply(get_grade)

print("=== 등급 추가 완료 ===")
df_clean
````

---

## 📌 Step 5. 조건 필터링 (원하는 데이터만 선택)

> 엑셀의 '필터' 기능처럼 조건에 맞는 행만 골라낼 수 있습니다!

````python
# 조건 필터링: 평균이 85점 이상인 학생만
# df[조건] 형태로 사용합니다
high_scorers = df_clean[df_clean['평균'] >= 85]

print("=== 평균 85점 이상인 학생 ===")
high_scorers
````

````python
# 여러 조건 조합
# & (AND) : 두 조건 모두 만족
# | (OR)  : 두 조건 중 하나라도 만족

# 예시 1: 1학년이면서 평균 80점 이상인 학생
condition = (df_clean['학년'] == 1) & (df_clean['평균'] >= 80)
result1 = df_clean[condition]
print("=== 1학년 & 평균 80점 이상 ===")
print(result1)
print()

# 예시 2: 수학 또는 영어가 90점 이상인 학생
condition2 = (df_clean['수학'] >= 90) | (df_clean['영어'] >= 90)
result2 = df_clean[condition2]
print("=== 수학 OR 영어 90점 이상 ===")
result2
````

````python
# 특정 값 포함 여부: isin()
# 'A' 또는 'B' 등급인 학생만 선택
top_grade = df_clean[df_clean['등급'].isin(['A', 'B'])]
print("=== A, B 등급 학생 ===")
top_grade
````

---

## 📌 Step 6. 정렬 (sort_values)

> 데이터를 원하는 기준으로 정렬합니다.

````python
# 평균 점수로 내림차순 정렬 (높은 점수가 위)
# ascending=False → 내림차순 (높은 순)
# ascending=True  → 오름차순 (낮은 순, 기본값)
df_sorted = df_clean.sort_values('평균', ascending=False)

print("=== 평균 점수 순위 (내림차순) ===")
df_sorted.reset_index(drop=True)
````

````python
# 여러 기준으로 정렬: 학년 오름차순, 그 안에서 평균 내림차순
df_sorted2 = df_clean.sort_values(['학년', '평균'], ascending=[True, False])
print("=== 학년순, 같은 학년이면 평균 높은 순 ===")
df_sorted2.reset_index(drop=True)
````

---

## 📌 Step 7. groupby - 그룹별 집계

> **groupby = 엑셀의 '피벗 테이블'**  
> 특정 기준으로 데이터를 묶어서 통계를 낼 때 사용합니다!

```
예시: 학년별 평균 점수가 궁금할 때
→ '학년'으로 그룹을 묶고 → 각 그룹의 '평균'을 구한다
```

````python
# 학년별 평균 점수
grade_avg = df_clean.groupby('학년')[['국어', '수학', '영어', '과학', '평균']].mean().round(1)

print("=== 학년별 과목 평균 ===")
grade_avg
````

````python
# 학년별 학생 수 세기
student_count = df_clean.groupby('학년').size()
print("=== 학년별 학생 수 ===")
print(student_count)
````

````python
# 등급별 통계 (agg로 여러 통계 한번에)
grade_stats = df_clean.groupby('등급')['평균'].agg(['count', 'mean', 'min', 'max'])
grade_stats.columns = ['인원수', '평균', '최저점', '최고점']

print("=== 등급별 통계 ===")
grade_stats.round(1)
````

## 📌 Step 8. 반도체 데이터 EDA 실습

> 실제 산업 데이터와 유사한 반도체 공정 데이터로 EDA를 실습합니다!

````python
# 반도체 공정 센서 데이터 생성 (100개 샘플)
np.random.seed(42)  # 재현 가능한 랜덤값 설정
n = 100

semi_data = {
    '공정ID':     [f'LOT{str(i).zfill(3)}' for i in range(1, n+1)],
    '온도(°C)':   np.round(np.random.normal(300, 15, n), 1),   # 평균 300도, 표준편차 15
    '습도(%)':    np.round(np.random.normal(45, 5, n), 1),     # 평균 45%, 표준편차 5
    '압력(Pa)':   np.round(np.random.normal(101.3, 2, n), 2),  # 평균 101.3Pa
    '가스유량(sccm)': np.round(np.random.normal(100, 8, n), 1), 
    '두께(nm)':   np.round(np.random.normal(50, 3, n), 2),     # 목표 두께 50nm
    '불량여부':   np.random.choice([0, 1], n, p=[0.85, 0.15])  # 성공(0): 85%, 실패(1): 15%
}

# 일부러 결측값 삽입
semi_data['온도(°C)'][5] = np.nan
semi_data['온도(°C)'][23] = np.nan
semi_data['습도(%)'][11] = np.nan
semi_data['두께(nm)'][45] = np.nan

df_semi = pd.DataFrame(semi_data)

print("=== 반도체 공정 데이터 (처음 10행) ===")
df_semi.head(10)
````

````python
# 기본 정보 확인
print("📊 데이터 기본 정보")
print("="*40)
df_semi.info()
print()
print("\n결측값 현황:")
print(df_semi.isnull().sum())
````

````python
# 통계 요약
print("📊 수치 데이터 통계 요약")
df_semi.describe().round(2)
````

````python
# 불량 여부별 통계 비교
result_stats = df_semi.groupby('불량여부')[['온도(°C)', '습도(%)', '압력(Pa)', '두께(nm)']].mean().round(2)
result_stats.index = ['✅ 정상(0)', '❌ 불량(1)']

print("=== 정상 vs 불량 공정 센서 평균값 비교 ===")
result_stats
````

````python
# 불량률 계산
total = len(df_semi)
defect_count = df_semi['불량여부'].sum()
defect_rate = defect_count / total * 100

print(f"총 공정 수: {total}개")
print(f"불량 발생: {defect_count}개")
print(f"불량률: {defect_rate:.1f}%")
````

## 📚 2차시 핵심 정리

| 기능 | 코드 | 설명 |
|------|------|------|
| **결측값 확인** | `df.isnull().sum()` | 열별 결측값 개수 |
| **결측값 삭제** | `df.dropna()` | 결측값 있는 행 삭제 |
| **결측값 채우기** | `df.fillna(값)` | 특정 값으로 대체 |
| **중복 확인** | `df.duplicated()` | 중복 행 여부 |
| **중복 제거** | `df.drop_duplicates()` | 중복 행 제거 |
| **새 열 추가** | `df['새열'] = 계산식` | 계산된 새 열 추가 |
| **함수 적용** | `df['열'].apply(함수)` | 함수를 각 값에 적용 |
| **조건 필터** | `df[df['열'] > 값]` | 조건에 맞는 행만 |
| **정렬** | `df.sort_values('열')` | 특정 열 기준 정렬 |
| **그룹별 집계** | `df.groupby('열').mean()` | 그룹별 평균 |
