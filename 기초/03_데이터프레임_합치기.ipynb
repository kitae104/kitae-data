### 데이터프레임 합치기
- 가로로 합치기 : df = pd.merge(df1, df2, how='left', on='id')
- 세로로 합치기 : df = pd.concat([df1, df2], ignore_index=False)



```python
import pandas as pd

df1 = pd.DataFrame({'id': [1, 2, 3, 4, 5], 'midterm': [60, 80, 70, 90, 85]})
print(df1)
```

```python
df2 = pd.DataFrame({'id': [1, 2, 3, 4, 5], 'final': [70, 83, 65, 95, 80]})
print(df2)
```

```python
df = pd.merge(df1, df2, how='left', on='id')
print(df)
```

```python
df_1 = pd.DataFrame({'id': [1, 2, 3, 4, 5], 'midterm': [60, 80, 70, 90, 85]})
df_2 = pd.DataFrame({'id': [6, 7, 8, 9, 10], 'midterm': [70, 83, 65, 95, 80]})
df_3 = pd.concat([df_1, df_2], ignore_index=True) # 세로로 합치기
print(df_3)
```
