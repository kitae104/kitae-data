### 조건에 맞는 데이터만 검색(필터링)

- 비교연산자:

- 논리연산자:
    - and, or, not
- 비트연산자:
    - &, |, ~
- 포함연산자:
    - in, not in

```python
import pandas as pd
df = pd.read_csv('exam.csv', header=0)
print(df)
```

```python
df1 = df.query('nclass==1') # 1반인 경우 추출
print(df1)
```

```python
df1 = df.query('nclass!=1') # 1반이 아닌 경우 추출
print(df1)
```

```python
df1 = df.query('math>80 and nclass!=1') # 수학 점수가 80점 이상이고 1반이 아닌 경우 추출
print(df1)
```

```python
# 영어 점수가 50점 이상이고 수학 점수가 50점 이상이며 1반인 경우 추출
df1 = df.query('eng>=50 and math>=50 and nclass==1')
print(df1)
```

```python
# 1반, 3반, 5반인 경우 추출
df1 = df.query('nclass==1 or nclass==3 or nclass==5')
print(df1)
```

```python
# 1반이면서 test가 Pass인 경우 추출
df1 = df.query('nclass ==1 and test == "Pass"')
df1
```

### 새로운 컬럼 추가하기
- 파생 컬럼(변수) 추가하기 
    - df['avg'] = df['tot'] / 3
- assign() 함수(C5블록)
    - df = df.assign(avg = df['tot'] / 3)
    - df = df.assign(avg = df['tot'] / 3, me_avg = (df['math']+df['eng'])/2)
    - df = df.assign(retest = np.where(df['avg'] >= 60, 'No', 'Ok'))

```python
import pandas as pd
df = pd.read_csv('exam.csv', header=0)
df['avg'] = df['tot'] / 3
print(df)
```

```python
df = df.assign(avg = df['tot'] / 3)
print(df)
```

```python
# 여러 개의 컬럼을 한 번에 추가하기
df = df.assign(avg = df['tot'] / 3, me_avg = (df['math']+df['eng'])/2)
print(df)
```

```python
import numpy as np
df = df.assign(retest = np.where(df['avg'] >= 60, 'No', 'Ok'))
print(df)
```

### 집단(그룹)별 요약 통계, 통계 요약(Summary Statistics)

1. df.groupby('컬럼명') : 지정 컬럼을 기준으로 그룹핑(집단화)
    - df.groupby('nclass')
    - df.groupby('nclass').mean()

### 요약통계함수( sum(), count(), max(), min(), median(), std() )
2. df.agg('컬럼명', '요약통계함수') : 다중 집계(aggregation) 작업
    - df.agg(mean_math = ('math', 'mean'))
    - df.agg(mean_math = ('math', 'mean'), max_math = ('math', 'max'))

```python
import pandas as pd
df = pd.read_csv('exam.csv', header=0)
# print(df)
# 숫자형으로 변환할 컬럼 목록 지정 (필요시 컬럼명 수정)
num_cols = ['math', 'eng', 'science', 'tot']
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors='coerce')
print(df.dtypes)
print(df.groupby('nclass')[num_cols].mean()) # nclass별 평균 구하기
print(df.groupby('nclass')[num_cols].mean()['math']) # nclass별 math 평균 구하기
print(df.groupby('nclass')[num_cols].max()['math']) # nclass별 math 최대값 구하기
```

```python
print(df.groupby('nclass')[num_cols].agg(['mean', 'max'])) # nclass별 math 평균과 최대값 구하기
```

```python
print(df.groupby('nclass')[num_cols].agg(mean_math = ('math', 'mean'), max_math = ('math', 'max'))) # nclass별 math 평균과 최대값 구하기
```

```python
print(df.groupby('nclass', as_index=False)[num_cols].agg(mean_math = ('math', 'mean'), max_math = ('math', 'max'))) # nclass별 math 평균과 최대값 구하기
```

### 계층적 집단(그룹)별 요약 통계, 통계 요약(Summary Statistics)
1. df.groupby( ['1차그룹컬럼명', '2차그룹컬럼명'] )
    - df.groupby( ['manufacturer', 'drv'] ) # 제조사별 구동방식별 그룹핑
    - df.groupby( ['manufacturer', 'drv'] ).mean()
    
### 요약통계함수( sum(), count(), max(), min(), median(), std() )
2. df.agg('컬럼명', '요약통계함수') : 다중 집계(aggregation) 작업
    - df.agg(mean_cty = ('cty', 'mean'))
    - df.agg(mean_cty = ('cty', 'mean'), max_cty = ('cty', 'max'))

```python
import pandas as pd
df = pd.read_csv('mpg.csv', header=0)
print(df.head())
```

```python
print(df.groupby(['manufacturer', 'drv']).mean(numeric_only=True)) # manufacturer와 drv별 평균 구하기
```

```python
print(df.groupby(['manufacturer', 'drv'])[[ 'cty', 'hwy']].mean(numeric_only=True)) # manufacturer와 drv별 평균 구하기
```

```python
# manufacturer와 drv별 cty 평균과 최대값 구하기
print(df.groupby(['manufacturer', 'drv']).agg(mean_cty=('cty', 'mean'), max_cty=('cty', 'max'))) 
```
