### 통계 분석
1. 기술통계(Descriptive Statistics)
- 데이터를 요약해 설명하는 통계 분석 기법

2. 추론통계(Inferential Statistics)
- 단순히 숫자를 요약하는 것을 넘어, 어떤 값이 발생할 확률 계산

3. 예시 : 성별에 따라 급여 차이 분석
- 기술통계 분석 : 성별에 따른 평균 급여 계산으로 차이를 분석 (남성 100, 여성 93)
- 추론통계 분석 : 급여 차이가 유의미한 것인지 분석 (유의확률(p-value) : 우연히 나타날 확률로 판정)

4. 통계적 가설 검정 : 유의확률을 이용해 가설 검정하는 방법
- t-검정(t-test)
    - 유의확률 5%를 판단 기준(p-value < 0.05, 유의미)
- 상관관계(correlation)
    - 두 변수 사이에 관련성 분석(상관계수(0~1))

### 기술 통계 분석
[문제 1] 자동차 연비 데이터셋(mpg.csv)에서 자동차 종류가 compact, suv 자동차의 도시 연비(cty)를 분석하시오.

```python
import pandas as pd
df = pd.read_csv('mpg.csv', header=0) # header=0은 첫 번째 행이 열 이름임을 나타냅니다.
df.head()
```

```python
# 조건 검색
from numpy import mean


df_stat = df.query('category in ["compact", "suv"]') # category 열에서 compact 또는 suv인 행을 선택하여 df_stat에 저장
df_stat
```

```python
df_stat = df.query('category in ["compact", "suv"]').groupby('category', as_index=False).agg({'category': 'count', 'cty': 'mean'}) # category 열에서 compact 또는 suv인 행을 선택하여 category별로 그룹화한 후, 각 그룹의 평균을 계산하여 df_stat에 저장
df_stat
```

```python
df_stat = df.query('category in ["compact", "suv"]').groupby('category', as_index=False).agg(n=('category', 'count'), mean=('cty', 'mean')) # category 열에서 compact 또는 suv인 행을 선택하여 category별로 그룹화한 후, 각 그룹의 평균을 계산하여 df_stat에 저장
df_stat
```

### 추론 통계, t-검정, t-test

```python
compact = df.query('category == "compact"')['cty'] # category 열에서 compact인 행을 선택하여 cty 열의 값을 compact에 저장
compact
```

```python
suv = df.query('category == "suv"')['cty'] # category 열에서 suv인 행을 선택하여 cty 열의 값을 suv에 저장
suv
```

```python
import scipy.stats as stats
t_statistic, p_value = stats.ttest_ind(compact, suv) # compact와 suv의 cty 열의 값을 사용하여 독립 표본 t-검정을 수행하여 t-통계량과 p-값을 계산하여 t_statistic과 p_value에 저장
print(f"t-통계량: {t_statistic}, p-값: {p_value}")
```

[문제 2] mpg.csv에서 일반 휘발유(regular)와 고급 휘발유(premium)의 도시 연비(cty)를 분석하시오.

```python
df_stat2 = df.query('fl in ["r", "p"]').groupby('fl', as_index=False).agg(n=('fl', 'count'), mean=('cty', 'mean')) # fl 열에서 "r" 또는 "p"인 행을 선택하여 fl별로 그룹화한 후, 각 그룹의 평균을 계산하여 df_stat에 저장
df_stat2
```

```python
regular = df.query('fl == "r"')['cty'] # fl 열에서 "r"인 행을 선택하여 cty 열의 값을 regular에 저장
premium = df.query('fl == "p"')['cty'] # fl 열에서 "p"인 행을 선택하여 cty 열의 값을 premium에 저장
t_statistic2, p_value2 = stats.ttest_ind(regular, premium) # regular와 premium의 cty 열의 값을 사용하여 독립 표본 t-검정을 수행하여 t-통계량과 p-값을 계산하여 t_statistic2와 p_value2에 저장
print(f"t-통계량: {t_statistic2}, p-값: {p_value2}")
```

### 상관관계(correlation)
- 두 변수 사이에 관련성 분석(상관계수(0~1))
- 상관계수 0.7 이상 : 강한 양의 상관관계
- 상관계수 0.3 이상 : 약한 양의 상관관계
- 상관계수 0.3 미만 : 거의 상관관계 없음
- 상관계수 : 양수(정비례), 음수(반비례)


[문제 1] 미국 경제 데이터셋(economics.csv)에서 실업자 수(unemployment)와 개인소비지출(pce) 간에 상호 관련성이 있는지 분석
1. date : 날짜(연-월-일)
2. pce(Personal Consumption Expenditure) : 개인소비지출 물가지수
3. pop(POPulation, in thousands) : 인구(천명 단위)
4. psavert(Personal Saving Rate) : 개인저축율
5. uempmed(median duration of unemployment, in weeks) : 실업지속기간(주 단위)
6. unemployment(number of unemployed in thousands) : 실업자 수(천명 단위)

```python
import pandas as pd
df = pd.read_csv('economics.csv', header=0) # header=0은 첫 번째 행이 열 이름임을 나타냅니다.
df.head()
```

```python
# pce와 unemploy 만 추출해서 확인
df1 =  df[['pce', 'unemploy']]
df1
```

```python
# 피어슨 상관계수 계산
df1.corr(method='pearson') # df1의 pce와 unemploy 열의 상관계수를 계산하여 출력합니다. method='pearson'은 피어슨 상관계수를 사용하여 계산하도록 지정합니다.
```

```python
import scipy.stats as stats
corr1 = stats.pearsonr(df1['pce'], df1['unemploy']) # df1의 pce와 unemploy 열의 피어슨 상관계수와 p-값을 계산하여 출력합니다. stats.pearsonr 함수는 두 개의 배열을 입력으로 받아 피어슨 상관계수와 p-값을 반환합니다.
print(f"피어슨 상관계수: {corr1[0]}, p-값: {corr1[1]}")
```
