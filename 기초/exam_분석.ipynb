### 1 데이터 불러오기 & 기본 탐색
- CSV 파일을 DataFrame으로 로드하고 기본 구조를 확인합니다

1-1. 라이브러리 임포트 & 데이터 불러오기
1. 판다스 분석에 필요한 라이브러리를 먼저 불러옵니다. 
2. pd.read_csv()로 CSV 파일을 DataFrame으로 읽어옵니다.

```python
import pandas as pd

# CSV 파일 불러오기
df = pd.read_csv('exam.csv')  # C1 블록(다양한 형식 불러오기)
# print(df)
df.head(3)  # C2 블록(데이터 확인하기)
```

1-2. 데이터 크기와 구조 확인
- 데이터를 불러온 뒤 가장 먼저 크기, 열 정보, 데이터 타입을 확인합니다.

```python
# 행 x 열 개수 확인
print("데이터 크기:", df.shape)  # C7 (데이터 크기 확인하기)

# 열 이름 확인

print("열 이름:", df.columns.tolist()) # C7 (열 이름 확인하기)

# 데이터 타입 및 결측치 확인
print(df.info()) # C2 블록(다양한 데이터 정보 확인하기)
```

1-3. 앞/뒤 데이터 미리보기

```python
# 앞에서 5행
print(df.head()) # C2 (데이터 확인하기) 

# 뒤에서 5행
print(df.tail())

# 특정 열만 보기 - 학생 번호와 과목 점수만
print(df[['no', 'math', 'eng', 'science']].head()) # 특정 컬럼만 [[컬럼명1, 컬럼명2]] 형태로 리스트로 전달
```

### 2.기술통계로 데이터 이해하기
- 평균, 최대/최솟값 등 통계량으로 데이터의 전체 경향을 파악합니다

2-1. 전체 통계 요약
- df.describe()는 숫자형 열에 대해 개수, 평균, 표준편차, 최솟/최댓값, 사분위수를 한 번에 보여줍니다.

```python
print(df.describe()) # C2 (기술통계로 데이터 이해하기)
```

2-2. 개별 통계 함수
- 특정 열이나 통계량만 따로 확인할 때 사용합니다.

```python
df_math = df['math']    # 특정 열 선택하기 - 수학 점수만
print(df_math)

df_mes = df[['math', 'eng', 'science']]  # 여러 열 선택하기 - 수학, 영어, 과학 점수
print(df_mes)
```

```python
# 과목별 평균
df_math_mean = df['math'].mean()  # C8 (평균 계산하기)
print("수학 점수 평균:", df_math_mean)
print("수학 점수 평균:", df_math_mean.round(1))  # 소수점 한 자리로 반올림
```

```python
print("영어 평균:", df['eng'].mean().round(1))
print("과학 평균:", df['science'].mean().round(1))
```

```python
# 합계 점수 최대·최솟값
print("최고 합계:", df['tot'].max()) # C8 (최대값 계산하기)
print("최저 합계:", df['tot'].min()) # C8 (최소값 계산하기)
```

```python
# 공부 시간 중앙값 (중간값)
print("공부 시간 중앙값:", df['hours'].median()) # C8 (중앙값 계산하기)
```

```python
# 여러 열 평균 한 번에
print(df[['math', 'eng', 'science']].mean().round(1))
```

2-3. 합격/불합격 비율 확인
- 범주형(문자) 열은 value_counts()로 빈도를 확인합니다.

```python
# 합격/불합격 빈도 수
print(df['test'].value_counts()) # C8 (범주형 데이터 빈도 수 계산하기)

# 비율로 보기 (normalize=True)
print(df['test'].value_counts(normalize=True).round(2)) 
```
