### 데이터 정제(clearning) - 결측치, 이상치 처리
1. 결측치(missing value)
- 누락된 값, 비어 있는 값
- 데이터 수집 과정에서 발생한 오류로 포함될 가능성
- 함수가 적용되지 않거나 분석 결과가 왜곡되는 문제 발생
- 실제 데이터 분석 시 결측치 확인, 제거 후 분석해야 함
- 결측치는 NaN(Not a Number)으로 표시됨
- 결측치 확인 : df.isna()
- 결측치 제거 : df.dropna()

2. 결측치 대체법(imputation)
- 결측치를 제거하는 대신 다른 값을 채워 넣는 방법
- 대표값(평균값, 최빈값, 중앙값 등)을 구해 일괄 대체
- 통계 분석 기법으로 결측치의 예측값 추정 후 대체
- 결측치 대체 : df.fillna()

```python
import pandas as pd
df = pd.read_csv('exam_nan.csv', header=0)
print(df)
```

```python
print(df.isna()) # 결측치 여부 확인
```

```python
print(df.isna().sum()) # 결측치 개수 확인
```

```python
# 결측치 제거하기
df1 = df.dropna(subset=['math', 'eng', 'science'])
print(df1)
```

```python
print(df['math'].mean()) # math 열의 평균값 계산 (결측치 포함)
print(df['math'].mean(skipna=True)) # math 열의 평균값 계산 (결측치 제외)
```

```python
df['math'] = df['math'].fillna(55)
print(df)

df['eng'] = df['eng'].fillna(df['eng'].mean())
print(df)
```
